```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("Financial_Sample.csv")
df.head()
```

```python
df.info()

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

# Check nulls
df.isnull().sum()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_units = df['Units Sold'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Total Units:", total_units)
```

```python
sales_trend = df.groupby('Date')['Sales'].sum().reset_index()

fig = px.line(sales_trend, x='Date', y='Sales', title='Sales Trend Over Time')
fig.show()
```

```python
country_sales = df.groupby('Country')['Sales'].sum().reset_index()

fig = px.bar(country_sales, x='Country', y='Sales', title='Sales by Country')
fig.show()
```

```python
fig = px.pie(df, names='Segment', values='Profit', title='Profit by Segment')
fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(x=country_sales['Country'], y=country_sales['Sales'], name='Sales'))
fig.add_trace(go.Scatter(x=sales_trend['Date'], y=sales_trend['Sales'], name='Trend'))

fig.update_layout(title='Business Dashboard')
fig.show()
```
